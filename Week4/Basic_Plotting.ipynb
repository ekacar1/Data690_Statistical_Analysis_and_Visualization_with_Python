```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from numpy import pi
```

```python
x = np.arange(-2*pi,2*pi,pi/8)
y = np.sin(x)
plt.ion()   # Turn interactive mode on
# Interactive = the plot gets updated as you go along
# For non-interactive, the plot doesn’t show up until you’ve finished everything
# to turn interactive mode off, enter the following command
# plt.ioff() 
plt.plot(x, y)
```

```python
# Change the line width:
plt.plot(x, y, linewidth=3)
```

```python
# to change the line style
plt.plot(x, y, linewidth=3, linestyle='dashed')
```

```python
# to change the line color
plt.plot(x, y, linewidth=3, linestyle='dashed', color='r')
```

```python
# to add point markers:
plt.plot(x, y, linewidth=3, linestyle='dashed', color='k', marker='d')
```

```python
# another marker
plt.plot(x, y, linewidth=3, linestyle='dashed', color='b', marker='*')
```

```python
# yet another marker
plt.plot(x, y, linewidth=3, linestyle='none', color='y', marker='s')
```

```python
# You can use LaTeX symbols as markers
plt.plot(x, y, linewidth=3, linestyle='none', color='m', marker='$\\beta$')
```

```python
# you can change the marker size
plt.plot(x, y, linewidth=3, linestyle='none', color='m', marker='$\\beta$', markersize = 10)
```

```python
# assume our y values have some error, shown by yerrrors. this error is random and let's say <= 0.4
yerrors = np.random.uniform(0,0.4,len(y))
```

```python
plt.errorbar(x, y, yerr=yerrors)
```

```python
# Creating error bars without lines:
plt.errorbar(x, y, yerr=yerrors, fmt='*')
```

```python
xx = np.arange(0,4.1,0.1)
```

```python
yy1 = 1 + np.cos(xx/4*pi) + np.exp(-xx) + 1/(1+(xx-2.05)**2)
yy2 = 1/(1+(xx-2.05)**2)
```

```python
plt.plot(xx,yy1)
```

```python
# PLOTTING MULTIPLE LINES and LABELING
plt.plot(xx,yy1, label='Function-1')
plt.plot(xx,yy2, label='Function-2')
plt.xlim([1, 3.6])
plt.xlabel('t (s)')
plt.ylabel('V (m/s)')
plt.title('Two Functions')
plt.legend(frameon=True)
```

```python
x = np.arange(-2*pi,2*pi,pi/20)
y = np.sin(x)
plt.plot(x,y)
```

```python
# SETTING AREA
# add_axes uses matplotlib’s location format, which takes the format of:
# [left, bottom, width, height] 
# where each of the numbers are from 0 to 1 (in units of a fraction of the figure)
fig1 = plt.figure()
ax1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
ax1.plot(x, y, marker='o', label='plotted line')
ax1.legend()
# All of those major plotting functions (i.e, plot, scatter, legend, et cetera) 
# are now just methods on the axis.
```

```python
ax1.set_xlim([1, 8])
ax1.set_ylim([-2, 3])
# All axis properties (i.e., x/ylim, x/yscale) can be set by the methods axis.set_property.
# Also, you can get the current values for these by axis.get_property.
```

```python
# try these
ax1.set_xscale('log')
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
```

```python
# Saving the figure is a method of the figure itself
fig1.savefig('plot2.png')
```
